```python
import numpy as np
import pandas as pd

data_airbnb= pd.read_csv('Airbnb_Open_Data.csv', sep=',', error_bad_lines=False, index_col=False, dtype='unicode')

```

## PREPROCESSING DATA

```python
data_airbnb.isnull().sum()
data_airbnb.a
```

```python
#mengecek duplikasi data

data_airbnb.duplicated().value_counts()
```

```python
#setelah dilakukan pengecekan duplikasi data. terdapat 541 yang mengandung duplikasi data, maka dilakukan penghapusan data.

data = data_airbnb.drop_duplicates(keep='first')
```

```python
data.isnull().sum().sort_values(ascending=True)

#data.dropna(axis=1)
```

```python
columns = ['NAME','license','house_rules','country','country code','reviews per month','last review']
data_airbnb.dropna('columns', axis = 1, inplace = True)
data.info()

```

```python
null_values = data['host_identity_verified'].isnull().sum()
null_values.sort_values()

total_cells=np.product(data['host_identity_verified'].shape)

total_null=null_values.sum()

#Now we have to calculate the percentage of null values
pct_null = (total_null/total_cells) * 100 
print('Total null values is : {} %'.format(pct_null))
```
